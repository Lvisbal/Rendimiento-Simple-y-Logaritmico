En el √°mbito de las finanzas, tanto el rendimiento simple como el rendimiento logar√≠tmico son medidas utilizadas para evaluar el desempe√±o o retorno de una inversi√≥n a lo largo del tiempo. Ambas medidas son importantes para comprender c√≥mo ha evolucionado el valor de una inversi√≥n y comparar diferentes activos o estrategias de inversi√≥n.

El rendimiento simple se calcula como el cambio absoluto en el valor de una inversi√≥n dividido por el valor inicial, expresado como un porcentaje. Es una medida lineal que no tiene en cuenta el efecto del inter√©s compuesto. La f√≥rmula del rendimiento simple se muestra a continuaci√≥n:

Rendimiento Simple = (Valor Final - Valor Inicial) / Valor Inicial

Por otro lado, el rendimiento logar√≠tmico (tambi√©n conocido como rendimiento logar√≠tmico continuo o tasa de rendimiento logar√≠tmica) tiene en cuenta el efecto del inter√©s compuesto y proporciona una medida m√°s precisa y adecuada para evaluar el rendimiento a largo plazo. Se calcula como el logaritmo natural del cociente entre el valor final y el valor inicial. La f√≥rmula del rendimiento logar√≠tmico se muestra a continuaci√≥n:

Rendimiento Logar√≠tmico = ln(Valor Final / Valor Inicial)

El rendimiento logar√≠tmico tiene la ventaja de que es aditivo en el tiempo, lo que significa que se pueden sumar los rendimientos logar√≠tmicos de diferentes per√≠odos para obtener el rendimiento acumulado total.

```python
#Importar Librerias
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
#Descargar los datos manteniendo solo los precios ajustados de cierre
df = yf.download('AAPL', start = '2020-01-01', end = '2023-05-31', progress=False)
df = df.loc[:, ['Adj Close']]
df.rename(columns={'Adj Close':'adj_close'}, inplace=True)
```

```python
#Calcular rendimientos simples y logaritmicos usando los precios ajustados de cierre
df['simple_rtn'] = df.adj_close.pct_change()
df['log_rtn'] =np.log(df.adj_close/df.adj_close.shift(1))
```

```python
#Importar librerias de graficos, agregar precios y rendimientos
import matplotlib.pyplot as plt
import warnings
fig, ax = plt.subplots(3, 1, figsize=(12, 10), sharex=True)
df.adj_close.plot(ax=ax[0])
ax[0].set(title = 'APPL Time Series', ylabel = 'Stock Prices ($)')
df.simple_rtn.plot(ax=ax[1])
ax[1].set(ylabel = 'Simple Return (%)')
df.log_rtn.plot(ax=ax[2])
ax[2].set(xlabel = 'Date', ylabel = 'Logaritmics Returns (%)')
ax[2].tick_params(axis='x', which='major', labelsize=12)
# plt.tight_layout()
# plt.savefig('Images/ch1_im7.png)
plt.show()
```

En resumen, el rendimiento simple es una medida lineal que no considera el efecto del inter√©s compuesto, mientras que el rendimiento logar√≠tmico tiene en cuenta el inter√©s compuesto y proporciona una medida m√°s precisa del rendimiento a largo plazo. Ambas medidas son utilizadas para evaluar y comparar inversiones.
